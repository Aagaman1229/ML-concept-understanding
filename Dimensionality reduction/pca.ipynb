PCA Learning

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
````

✅ Load a simple dataset

The Iris dataset is a classic dataset in machine learning. It contains:

150 samples of iris flowers

4 features (columns):

Sepal length

Sepal width

Petal length

Petal width

3 classes (types of flowers):

Setosa (0)

Versicolor (1)

Virginica (2)

````python
# ✅ Load a simple dataset
iris = load_iris()
X = iris.data  # only features
y = iris.target  # labels
feature_names = iris.feature_names
````

lets visualize the data

````python
# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target (species label) to the DataFrame
df['species'] = iris.target

# Display the first 10 rows
df.head(10)
````

dataset

````python
# ✅ Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ✅ Perform PCA using sklearn
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# ✅ Plot original vs PCA-reduced data
plt.figure(figsize=(14, 5))
````

original data

````python
# Original Data (first 2 features)
plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap='viridis')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title("Original Data (First 2 Features)")
````

pca data

````python

# PCA Reduced Data
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Transformed Data (2 Components)")
plt.grid(True)
plt.show()
````

### 🎯 PCA Result Interpretation

We plotted **PC1 vs PC2** using the `iris` dataset, with color (`c=y`) representing the 3 flower species:

| Label | Species     |
|-------|-------------|
| 0     | Setosa      |
| 1     | Versicolor  |
| 2     | Virginica   |

#### 🟢 PCA Plot Insights:

- **Setosa** is clearly separated from the other two species.  
- **Versicolor** and **Virginica** show some **overlap**, but still maintain partial separation.
  
> ✅ This shows that PCA did a good job of reducing dimensions **from 4D to 2D** while keeping **useful class separation**.

#### 📊 Why This Matters:

- PCA helps us **visualize high-dimensional data** in 2D.
- It preserves **maximum variance** in fewer dimensions.
- The clear separation of classes (especially Setosa) suggests PCA output can be useful for:
  - 🧠 Clustering  
  - 🤖 Classification  
  - 📉 Noise reduction

---

### 📈 Explained Variance Ratio

Example output:
```python
[0.72962445 0.22850762]

PC1 (Principal Component 1) explains ~72.77% of the variance.

PC2 explains ~23.03%.

Together, PC1 + PC2 = ~96% → This means your 2D PCA space preserves 96% of original info, which is excellent.

let's see the data

````python

# ✅ Show explained variance
print("Explained variance by each principal component:")
print(pca.explained_variance_ratio_)

# ✅ Manual PCA: Covariance Matrix, Eigenvalues, Eigenvectors
cov_matrix = np.cov(X_scaled.T)
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)

print("\nEigenvalues (Variance of each PC):")
print(eig_vals)

print("\nEigenvectors (Principal directions):")
print(eig_vecs)

# ✅ Plot explained variance
plt.figure(figsize=(6, 4))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, alpha=0.7, color='orange')
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Explained Variance per Principal Component")
plt.grid(True)
plt.show()
````
